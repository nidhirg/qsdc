```python
import matplotlib.pyplot as plt
import numpy as np
fact=lambda x : np.math.factorial(int(np.ceil(x)))
```

```python
def extra_error(gamma=0.05, epsilon=0.25, N=100, check=10, eve=10):
    if check + eve > N:
        return (1 - 2 * gamma) * epsilon
    return (1 - 2 * gamma) * epsilon * (1 - (fact(N - check) * fact(N - check - eve)) / (fact(N) * fact(N - eve)))
```

```python
gammas = np.linspace(0.01, 0.4)
for gamma in gammas:
    plt.scatter(gamma, extra_error(gamma=gamma), color='blue')
plt.show()
```

```python
eps = np.linspace(0.01, 0.98)
eps[-1] = 0.001
eps[-1] = 0.0001
for ep in eps:
    plt.scatter(ep, extra_error(epsilon=ep), color='blue')
plt.show()
```
